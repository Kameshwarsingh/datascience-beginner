```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
#jupyter notebook --NotebookApp.iopub_data_rate_limit=1.0e10
```

```python
#https://www.kaggle.com/datasets/darinhawley/fortune-500-companies-19552021?resource=download
df = pd.read_csv('fortune500_short.csv')
```

```python
#print(df.to_string()) 
```

```python
for col in df.columns:
    print(col)
```

```python
data_df=df[["Name","Revenue","Rank"]]
```

```python
#plt.plot(X-axis, Y-axis)
plt.rcParams["figure.figsize"] = (20,30)
plt.plot(data_df["Rank"],data_df["Revenue"])
plt.title("Scale - as is")
plt.grid()
plt.xticks(np.arange(1, 300, 5.0))
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (20,30)
plt.plot(data_df["Rank"],np.log(data_df["Revenue"]))
plt.title("Scale - log base e.")
plt.grid()
plt.xticks(np.arange(1, 300, 5.0))
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (20,30)
plt.plot(data_df["Rank"],np.log10(data_df["Revenue"]))
plt.title("Scale - log base 10.")
plt.grid()
plt.xticks(np.arange(1, 300, 5.0))
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (20,30)
plt.plot(data_df["Rank"],np.log2(data_df["Revenue"]))
plt.title("Scale - log base 2.")
plt.grid()
plt.xticks(np.arange(1, 300, 5.0))
plt.show()
```

```python
df = pd.read_csv('ford.csv')
data_df=df[["Year","Revenue"]]

data_df["Year"] = data_df["Year"].astype(int)
data_df["Revenue"] = data_df["Revenue"].astype(int)  
```

```python
plt.rcParams["figure.figsize"] = (30,30)
plt.plot(data_df["Year"],data_df["Revenue"])
plt.title("Scale - non.")
plt.grid()
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (30,30)
plt.plot(data_df["Year"],np.log10(data_df["Revenue"]))
plt.title("Scale - non.")
plt.grid()
plt.show()
```
