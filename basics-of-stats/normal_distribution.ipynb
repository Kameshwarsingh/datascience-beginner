```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creating a Function.
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
```

```python
# Creating a series of data of in range of 1-50.
x = np.linspace(0,1,200)
#Calculate mean and Standard deviation.
mean = np.mean(x)
std = np.std(x)
```

```python
print(f"Shape: {x.shape}")
print(f"mean: {mean}  , std: {std}")
```

```python

#Apply function to the data.
pdf = normal_dist(x,mean,std)
 
#Plotting the Results
plt.plot(x,pdf , color = 'red')
plt.xlabel('Data points')
plt.ylabel('Probability Density')

#plot mean, std
plt.axvline(x = mean, color = 'r',label = 'mean')
plt.axvline(x = mean+std, color = 'g',label = 'Std',alpha=0.1)
plt.axvline(x = mean-std, color = 'g',alpha=0.1)
```

```python
## Area under curve
from scipy.stats import norm
#The location (loc) keyword specifies the mean. 
#The scale (scale) keyword specifies the standard deviation.
```

```python
##Cumulative
cdf_lower_limit=norm(loc =mean , scale = std).cdf(0.4)
cdf_upper_limit =norm(loc =mean , scale = std).cdf(0.6)
```

```python
print(f"Lower-limit:{cdf_lower_limit}, Upper-limit:{cdf_upper_limit} , Difference:{cdf_upper_limit-cdf_lower_limit}")
```

```python
print("PPF/inverse of cdf  percentiles:",norm.ppf(0.634, loc=mean, scale=std))
#print("PDF:",norm.pdf(0.634, loc=mean, scale=std))
```
