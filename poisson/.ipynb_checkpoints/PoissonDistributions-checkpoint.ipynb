```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Set random seed for reproducibility
np.random.seed(42)

# Generate a sample dataset with a Poisson distribution
lambda_parameter = 3.5  # Average rate of events per interval
sample_size = 1000
sample_data = np.random.poisson(lambda_parameter, size=sample_size)

# Plot the histogram of the sample data
plt.hist(sample_data, bins=np.arange(0, max(sample_data) + 1) - 0.5, density=True, alpha=0.5, label='Sample Data')

# Plot the theoretical Poisson PMF
x_values = np.arange(0, max(sample_data) + 1)
poisson_pmf_values = poisson.pmf(x_values, lambda_parameter)
plt.plot(x_values, poisson_pmf_values, 'r-', label='Poisson Distribution (Î»={})'.format(lambda_parameter))

plt.title('Poisson Distribution Example')
plt.xlabel('Number of Events')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import poisson
from scipy.optimize import curve_fit

# Load your dataset
df = pd.read_csv('possion_dataset.csv')
```

```python
# Assuming 'event_count' is a column in your dataset representing the number of events
event_count_column = 'event_count'  # Replace with your actual column name
# Calculate the observed mean and standard deviation from the dataset
observed_mean = df[event_count_column].mean()
```

```python
observed_mean
```

```python
# Generate a sample dataset with a Poisson distribution
lambda_parameter = observed_mean # Average rate of events per interval
#sample_size = 200
#sample_data = np.random.poisson(lambda_parameter, size=sample_size)
sample_data=df[event_count_column]
```

```python
# Plot the histogram of the sample data
#plt.hist(sample_data, bins=np.arange(0, max(sample_data) + 1) - 0.5, density=True, alpha=0.5, label='Sample Data')
plt.hist(sample_data, bins=np.arange((min(sample_data)-10), max(sample_data) + 10), density=True, alpha=0.5, label='Sample Data')

# Plot the theoretical Poisson PMF
x_values = np.arange(0, max(sample_data) + 1)
poisson_pmf_values = poisson.pmf(x_values, lambda_parameter)
plt.plot(x_values, poisson_pmf_values, 'r-', label='Poisson Distribution (Î»={})'.format(lambda_parameter))

plt.title('Poisson Distribution Example')
plt.xlabel('Number of Events')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
