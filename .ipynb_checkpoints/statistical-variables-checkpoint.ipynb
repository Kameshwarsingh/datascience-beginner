#### Type of statistical variables
- Categorical (Qualitative/Boolean/Groups/Category/)
    - Nominal (No rank/order Example - Hair color)
    - Ordered (Rank/order Example - Qualitative/exam grades (A,B,C), Ranking of agree/degree (number of stars))
- Numeric (Quantitative)
    - Continuous (Real number, Example - Temperature of day, Balance in bank-account)
    - Discrete (Integer-Only, Example - Number of people in room, Number of rooms in a flat)

Note: Continuous and Discrete variables can further be divided into scales of measurement -  "Ratio scale" or "Interval scale " 

How do I summarize (graphically and numerically) categorical or qualitative variable?

- Frequency tables
- Bar charts
- Pie charts

How do I summarize numerically variable?
- Frequency tables
- Histogram
- Kernel density plot

- Histograms provide a binned representation of the data, highlighting the frequency or density of observations in each bin.

- Kernel Density Plots provide a smooth estimate of the probability density function, allowing for a more continuous representation of the distribution.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
# Generate some random data
data = np.random.randn(1000)
```

```python

# Create a kernel density plot
fig, axes = plt.subplots(2, 2, figsize=(18, 10))
sns.histplot(ax=axes[0, 0],data=data, kde=False, color='green', bins=50)
sns.kdeplot(ax=axes[0, 1],data=data, shade=False, color='red')
sns.histplot(ax=axes[1, 0],data=data, kde=True, color='blue', bins=50)
sns.boxplot(ax=axes[1, 1],data=data, color='green')
```
